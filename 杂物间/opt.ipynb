```python
###画出函数图像
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def f(x):
    return 3*(x**4)-16*(x**3)+30*(x**2)-24*x+8

# 生成 x 值的范围
x = np.linspace(0,3, 100)

# 计算 y 值
y = f(x)

# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = 3* \\alpha ^4 - 16* \\alpha ^3 + 30* \\alpha^2-24*\\alpha+8 $')
plt.grid(True)
plt.show()
```

```python
### 二分法求解
mid_lst = []
import numpy as np
def end(diff,a_k,b_k,lr = 100,jd = 0.0000001,time=0):
    mid_lst.append([a_k,b_k])
    mid = (a_k + b_k)/2
    vue = diff(mid)
    if vue==0:
        return mid
    elif abs(a_k-b_k)<jd or time==lr:
         return [a_k,b_k]
    elif np.sign(vue) == np.sign(diff(a_k)):
         return end(diff,mid,b_k,time=time+1)
    elif np.sign(vue) == np.sign(diff(b_k)):
         return end(diff,a_k,mid,time=time+1)
    else:
         print(vue)
         return 999
```

```python
### 三分法求解
mid_lst = []
import numpy as np
def sf_end(f,a_k,b_k):
    mid_lst.append([a_k,b_k])
    u_k = (b_k - a_k)*0.33+a_k
    v_k = (b_k - a_k)*0.66+a_k
    if f(u_k) == f(v_k):
       return sf_end(f,u_k,v_k)
    elif abs(a_k-b_k)<0.0000001:
         return [a_k,b_k]
    elif f(u_k) < f(v_k):
        return sf_end(f,a_k,v_k)
    elif f(u_k) > f(v_k):
        return sf_end(f,u_k,b_k)
    else:
        print("出错了")
```

```python
### 黄金分割法求解
import numpy as np
mid_lst = []
def hj_end(f,a_k,b_k,u=None,v=None):
    mid_lst.append([a_k,b_k])
    landa = (5**0.5-1)*0.5
    if  not u:
       u_k = a_k + (1-landa)*(b_k-a_k)
    else:
        u_k = u 
    if not v:  
        v_k = a_k + landa*(b_k-a_k)
    else:
        v_k = v
    if f(u_k) == f(v_k):
       return hj_end(f,u_k,v_k)
    elif abs(a_k-b_k)<0.001:
         return [a_k,b_k]
    elif f(u_k) < f(v_k):
        return hj_end(f,a_k,v_k,None,u_k)
    elif f(u_k) > f(v_k):
        return hj_end(f,u_k,b_k,u=v_k,v=None)
    else:
        print("出错了")
```
