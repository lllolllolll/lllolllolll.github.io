```python
### 
mid_lst = []
import numpy as np
def end(diff,a_k,b_k,lr = 100,jd = 0.0000001,time=0):
    mid_lst.append([a_k,b_k])
    mid = (a_k + b_k)/2
    vue = diff(mid)
    if vue==0:
        return mid
    elif abs(a_k-b_k)<jd or time==lr:
         return [a_k,b_k]
    elif np.sign(vue) == np.sign(diff(a_k)):
         return end(diff,mid,b_k,time=time+1)
    elif np.sign(vue) == np.sign(diff(b_k)):
         return end(diff,a_k,mid,time=time+1)
    else:
         print(vue)
         return 999
```

```python
### 
mid_lst = []
import numpy as np
def sf_end(f,a_k,b_k):
    mid_lst.append([a_k,b_k])
    u_k = (b_k - a_k)*0.33+a_k
    v_k = (b_k - a_k)*0.66+a_k
    if f(u_k) == f(v_k):
       return sf_end(f,u_k,v_k)
    elif abs(a_k-b_k)<0.0000001:
         return [a_k,b_k]
    elif f(u_k) < f(v_k):
        return sf_end(f,a_k,v_k)
    elif f(u_k) > f(v_k):
        return sf_end(f,u_k,b_k)
    else:
        print("")
```

```python
### 
import numpy as np
mid_lst = []
def hj_end(f,a_k,b_k,u=None,v=None):
    mid_lst.append([a_k,b_k])
    landa = (5**0.5-1)*0.5
    if  not u:
       u_k = a_k + (1-landa)*(b_k-a_k)
    else:
        u_k = u 
    if not v:  
        v_k = a_k + landa*(b_k-a_k)
    else:
        v_k = v
    if f(u_k) == f(v_k):
       return hj_end(f,u_k,v_k)
    elif abs(a_k-b_k)<0.001:
         return [a_k,b_k]
    elif f(u_k) < f(v_k):
        return hj_end(f,a_k,v_k,None,u_k)
    elif f(u_k) > f(v_k):
        return hj_end(f,u_k,b_k,u=v_k,v=None)
    else:
        print("")
```
