3根据你下载的机场历史数据，绘制该机场一天24小时的起降架次箱线图，横坐标是00:00-00:59，01:00-01:59，...., 23.00-23:59，纵坐标是起降架
次，将各小时箱线图的中位数连接起来作起降架次折线图。计算:高峰月日平均起峰架次(ADPM，Average Daily Peak Month) 和ADPM集中率计算
公式: ADPM/年日均起降架次) 。

```python
plt_lst = []
labels = []
for i in range(24):
    hour = str(i).zfill(2)
    hour_date = np.array(res.xs(hour,level=1))
    plt_lst.append(hour_date)
    labels.append(str(i)+":"+"00"+"-"+str(i)+":"+"59")
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# 设置图表标题
plt.title('起降架次箱线图')


# 批量绘制箱线图
plt.boxplot(plt_lst, labels=labels)
plt.xticks(rotation=90)
# 显示图表
plt.show()
```

```python
plt_med = []
for i in plt_lst:
    plt_med.append(np.median(i))
```

```python
import matplotlib.pyplot as plt

# 示例数据
x = labels
y = plt_med

# 绘制折线图
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('起降架次/小时中位数')
plt.ylabel('起降架次')

# 设置横轴标签横置，纵轴标签竖着展示
plt.xticks(rotation=90)

plt.show()
```
