```python
import pandas as pd
from docx import Document
# 创建一个新的Word文档
doc = Document()
for yr in yrs:
    dd1 = d1.groupby(["年","贸易国"])["贸易互补系数"].mean()[yr]
    dd2 = d2.groupby(["年","贸易国"])["贸易互补系数"].mean()[yr]
    dd3 = d3.groupby(["年","贸易国"])["贸易互补系数"].mean()[yr]
    dd4 = d4.groupby(["年","贸易国"])["贸易互补系数"].mean()[yr]
    index_new = list(set(d3.groupby(["年","贸易国"])["贸易互补系数"].mean()[yr].index).intersection(set(index)))
    df = pd.concat([dd1[index_new],dd2[index_new],dd3[index_new],dd4[index_new]],axis=1)
        
    df.columns = ["hs85","hs84","hs83","hs90"]
    df.loc["东盟"] = [dd1.mean(),dd2.mean(),dd3.mean(),dd4.mean()]
    df = df.round(4)
    # 插入标题
    doc.add_heading('yr'+str(yr), level=1)

    # 将DataFrame转换为表格
    table = doc.add_table(df.shape[0]+1, df.shape[1]+1)
    c = ["country",'hs85', 'hs84', 'hs83', 'hs90']
    # 添加表头
    for i, column_name in enumerate(c):
        table.cell(0, i).text = column_name

    # 添加数据
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            table.cell(i+1, 0).text = str(df.index[i])
            table.cell(i+1, j+1).text = str(df.values[i, j])
    # # 保存文档
doc.save(r'C:\Users\lll23\Desktop\应用\数据分析\实战\mid\date_js\output_v4.docx')
```
